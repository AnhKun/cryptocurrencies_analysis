# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

# Import data

```python
data = pd.read_csv('cc_ohlcv_by_exchange_landing_202109130813.csv')
data
```

# Assess data

```python
df = data.copy()
```

```python
df.info()
```

#### Note:
- missing values for column 'volume' and 'amount'



# Clean data
### Drop missing values for column 'volume'
I do not have any precise values to substitute the missing values for those rows in column 'volume'. Besides that, there are only 2 missing rows, so I just delete those rows.

```python
# check the rows with missing values
df[df['volume'].isnull()]
```

```python
# delete those rows
df = df[df['volume'].notnull()]
# test
df[df['volume'].isnull()]
```

### Substitute the rows with missing values in column 'amount'
By observing the dataset, the amount follows the equation: 
`amount = volume/price`

```python
df[df['amount'].isnull()]
```

The missing values occur because those rows have 0 in price. So I just substitute 0 for those missing values.

```python
# use fillna() to substitute 0 with missing values
df['amount'].fillna(0, inplace=True)
# test
df[df['amount'].isnull()]
```

# Analyze data

```python
new_df = df.groupby(['symbol', 'date']).mean()
new_df
```

```python
df[df['symbol']=='zrx-usdt']
```

```python
symbol = set(df['symbol'])
```

```python
new_df.loc['1inch-usd',:].price.index
```

```python
for i in symbol:
    plt.figure(figsize=[6,2])
    plt.plot(new_df.loc[i,:].price.index, new_df.loc[i,:].price)
    plt.ylabel('Price in $')
    plt.xticks(['2019-01-01', '2020-01-01','2021-01-01', '2022-01-01'], ['2019','2020','2021', '2022'])
    plt.xlabel('Year')
    plt.title(i)
    plt.show()
```
